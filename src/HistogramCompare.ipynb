```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import itertools

def calc_hist(img):
    hist = cv2.calcHist([img], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
    hist = cv2.normalize(hist, hist, 0, 255, cv2.NORM_MINMAX)
    hist = hist.squeeze(axis=-1)

    return hist

def plot_hist(hist, title=None):
    fig, ax = plt.subplots()
    if title:
        ax.set_title(title)
    ax.set_xticks([0, 255])
    ax.set_xlim([0, 255])
    ax.set_xlabel("Pixel Value")

    bins = np.arange(257)
    centers = (bins[:-1] + bins[1:]) / 2
    widths = np.diff(bins)
    ax.bar(centers, hist, width=widths, color="k")

    plt.show()

filename = '../../medicaldata/images/0001.jpg'
img = cv2.imread(filename,cv2.IMREAD_GRAYSCALE)
img_hist, img_bins = np.histogram(np.array(img).flatten(), bins=np.arange(256+1))

plt.plot(img_hist)
plt.show()
```
